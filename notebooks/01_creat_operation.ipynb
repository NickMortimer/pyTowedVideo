# Introduction

The Cannon Cameras have a number of usefull exif tags that will help us process the images with more certainty. The camera owner name should be set using the EOS Utility Software to one of the following option:

| Camera Name/Function | File Name Abbreviation |
| --- | --- |
|Survey Camera|SVY|
|Look Ahead Camera|LOOK|
|Survey Port Camera|SVYP|
|Survey Starboard Camera|SVYS|
|Stereo Camera Port|SCP|
|Stereo Camera Starboard|SCS|

Secondly a note of the BodySerialNumber should be made. Each time an operation is done images should be saved to a folder

```bash
capture_path
├───operation.json 
├───camera1  
|   └───img_0001.jpg|
└───camera2 
    └───img_0001.jpg 
```   
The operation is describe by a json file in the root directory of the capture directory 

## Example of json operation file 
{"message":"project.metadata","data":{"Voyage":"IN2018_V06"},"timestamp":"2015-11-05T07:00:26.860Z"}
{"message":"operation.metadata","data":{"Operation_identifier":"31","Start_time":"2015-11-05T07:00:26.860Z","End_time":"2015-11-05T10:00:26.860Z","Platform":"DTVC","Site_identifier":"VSM02"},"timestamp":"2015-11-05T07:00:26.860Z"}
{"message":"dtc.cam.scs.metadata","data":{"Camera_name":"SVY","Body_serial_number":"271011000031","Description":"Starboard stereo camera"},"timestamp":"2015-11-05T07:00:26.860Z"}
{"message":"dtc.cam.scp.metadata","data":{"Camera_name":"SCP","Body_serial_number":"271011000031","Description":"Port stereo camera"},"timestamp":"2015-11-05T07:00:26.860Z"}

# Rename video and move

````python
def rename_video(item):
    item.new_file = item.Platform + '_SVY_' + item.Voyage + '_'+ \
                        str(item.Operation_identifier).zfill(3) + '_'+ item['QuickTime:TrackCreateDate'].strftime('%Y%m%dT%H%M%SZ')+\
                        os.path.splitext(os.path.split(item.SourceFile)[1])[1]
    if os.path.exists(item.origonal_file):
        print('Moving '+item.new_file)
        shutil.move(item.origonal_file,videodestination+item.new_file)
        # shutil.move(item['File:Directory']+item.new_file,videodestination+item.new_file)
    return(item)


metadata['Voyage'] = log.deployment_metadata['Voyage'] 
metadata['new_file'] =''
metadata['Platform'] = log.deployment_metadata['Platform']  
metadata['Operation_identifier'] = log.deployment_metadata['Operation_identifier']  
metadata =metadata.apply(rename_video,axis=1)

````
