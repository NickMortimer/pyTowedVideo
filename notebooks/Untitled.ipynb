```python
import pandas as pd
import glob
```

```python
video = glob.glob('S:/hsv/hsv_video*.csv')
```

```python
videos = pd.concat([pd.read_csv(vid,index_col='TimeStamp',parse_dates=['TimeStamp']) for vid in video])
```

```python
videos.sort_index(inplace=True)
v=videos.reset_index().drop_duplicates(subset='TimeStamp',keep='first').set_index('TimeStamp')
v['Camera_name'] ='SVY'
v['Voyage'] ='LI2019_V05'
v['Operation'] =1
v['Operation'] = v['Operation'].cumsum()
v['Platfrom'] = 'HSV'
v['VideoNumber'] =1
v['OldPath'] = 'E:/HSV/Video/'+v['File']
```

```python
v['FileNameTime'] =pd.to_datetime(v.TimeStampMeta).apply( lambda x:x.strftime('%Y%m%dT%H%M%SZ'))
```

```python
v['Site']=v['File'].str.split('-',expand=True)[0]
```

```python
v['NewFileName']=v.apply( lambda x:f'HSV_SVY_LI2019_V05_{x.Operation:03}_{x.FileNameTime}_{x.VideoNumber:04}.mp4',axis=1)
```

```python
v['Operation'] = v['Operation']-1
```

```python
v['NewPath'] =v.apply( lambda x:f'E:/LI2019_V05/HSV/LI2019_V05_{x.Operation:03}/HSV_SVY_LI2019_V05_{x.Operation:03}_{x.FileNameTime}_{x.VideoNumber:04}.mp4',axis=1)
```

```python
v['VideoEndTime'] =v.index +pd.to_timedelta(v['Duration'],'S')
v['VideoStartTime'] =v.index
```

```python
v[['Camera_name', 'Duration', 'File', 'NewPath', 'OldPath', 'Path', 'Platform', 'TimeStampMeta', 'Voyage',
       'image_count', 'Operation', 'Platfrom', 'VideoNumber','FileNameTime', 'Site', 'NewFileName', 'VideoEndTime','VideoStartTime']].to_csv('LI2019_V05_HSV_VIDEO.CSV')
```

```python
v.to_csv('LI2019_V05_HSV_VIDEO.CSV')
```
